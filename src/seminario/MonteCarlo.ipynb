```python
import numpy as np  # Librería para cálculos numéricos y generación de números aleatorios
import matplotlib.pyplot as plt  # Librería para generar gráficos


def f(x):
    return 40 + 1 - ((x ** 3 + x ** 2 - 60) / 900)


def monte_carlo_search(num_points, num_iterations=10):
    """
    Retorna:
        x_min_values: Lista con los valores de x donde se encontró el mínimo en cada iteración.
        f_min_values: Lista con los valores mínimos de f(x) encontrados.
    """
    x_min_values = []  # Lista para almacenar valores de x donde f(x) es mínimo
    f_min_values = []  # Lista para almacenar los valores mínimos de f(x)

    for _ in range(num_iterations):
        x_samples = np.random.uniform(3, 100,  num_points)  # Genera num_points valores aleatorios de x en el intervalo [3,100]
        f_values = f(x_samples)  # Calcula f(x) para todos los valores generados

        min_idx = np.argmin(f_values)  # Encuentra el índice del menor valor de f(x)
        x_min_values.append(x_samples[min_idx])  # Guarda el x correspondiente al mínimo
        f_min_values.append(f_values[min_idx])  # Guarda el valor mínimo de f(x)

    return x_min_values, f_min_values


# Experimentos con 30 y 100 puntos
x_min_30, f_min_30 = monte_carlo_search(30)  # Ejecuta la búsqueda con 30 puntos por iteración
x_min_100, f_min_100 = monte_carlo_search(100)  # Ejecuta la búsqueda con 100 puntos por iteración

# Visualización de resultados con histogramas
plt.figure(figsize=(10, 5))  # Define el tamaño del gráfico
plt.hist(x_min_30, bins=10, alpha=0.5, label='30 puntos')  # Histograma para el experimento con 30 puntos
plt.hist(x_min_100, bins=10, alpha=0.5, label='100 puntos')  # Histograma para el experimento con 100 puntos
plt.xlabel('Valor de x mínimo encontrado')  # Etiqueta del eje X
plt.ylabel('Frecuencia')  # Etiqueta del eje Y
plt.legend()  # Muestra la leyenda
plt.title('Distribución de valores mínimos encontrados')  # Título del gráfico
plt.show()  # Muestra el gráfico

# Cálculo de estadísticas
print("Promedio de x mínimo con 30 puntos:",
      np.mean(x_min_30))  # Calcula y muestra el promedio de x mínimo con 30 puntos
print("Promedio de x mínimo con 100 puntos:",
      np.mean(x_min_100))  # Calcula y muestra el promedio de x mínimo con 100 puntos

```
